```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#membaca dataset
df = pd.read_csv('Datasets/Salary_Data.csv')

df.info()
```

```python
#mengekstrak variabel independen (X) dan variabel dependen (y)
X = df['YearsExperience'].values.reshape(-1,1)
y = df['Salary'].values.reshape(-1,1)
```

```python
#membuat objek regresi linier
model = LinearRegression()

#menyesuaikan model dengan data
model.fit(X, y)

#memprediksi nilai y berdasarkan X
y_pred = model.predict(X)
```

```python
#membuat persebaran data
plt.scatter(X, y)

#memplot garis regresi
plt.plot(X, y_pred, color='red')

#mengatur label sumbu dan judul
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Linear Regression')

#menampilkan plot dalam bentuk grafik dengan garis regressionnya
plt.show()
```
